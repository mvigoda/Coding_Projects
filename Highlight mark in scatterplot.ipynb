```python
import altair as alt
import pandas as pd


from altair_plot_utilities import *
```

```python


def point_in_scatterplot(data,small_data,X_axis_title,Y_axis_title):
    
    chart1 = alt.Chart(data).mark_point(filled=True, size=150).encode(
        x='x',
        y='y',
    color=alt.value('red'))
    
    
    chart2 = alt.Chart(small_data).mark_point(filled=True,  size=50).encode(
         x=alt.X('x:Q', axis=alt.Axis(title= X_axis_title)),
         y=alt.Y('y:Q', axis=alt.Axis(title= Y_axis_title )),
        color=alt.value('black')
    )
    return chart1 + chart2
```

```python
columns = ['x','y']
data = pd.DataFrame(columns = columns)
data

X = [int(1000*random.random()) for i in range(10)]
Y = [int(1000*random.random()) for i in range(10)]

data.x = X
data.y = Y
data

small_data = data.loc[0]
small_data  = pd.DataFrame(small_data).T
small_data
print(small_data)
```

```python
point_in_scatterplot(data,small_data, 'X_axis_title','Y_axis_title')
```

```python
from sklearn.datasets import make_blobs

X, labels = make_blobs(20, random_state=1)
points = pd.DataFrame({
    'x': X[:, 0],
    'y': X[:, 1],
    'labels': labels
})
centers = points.groupby('labels').mean()
data = pd.concat([points , centers.reset_index()])

chart1 = alt.Chart(data).mark_point(filled=True, size=150).encode(
    x='x',
    y='y',
    color='labels:N'
).properties(
    title = 'tidfdftle')

chart2 = alt.Chart(centers).mark_point(filled=True, size=50).encode(
    x='x',
    y='y',
    color=alt.value('black')
)

chart1 + chart2
```
